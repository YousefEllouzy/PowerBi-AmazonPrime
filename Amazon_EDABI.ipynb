```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('amazon_prime_titles.csv')
data
```

```python
data.columns
```

```python
data.info()
```

```python
data.duplicated().sum()
data=data.drop_duplicates()
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.drop(['description'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['director'].fillna('unvaliable',inplace=True)
data['cast'].fillna('unvaliable',inplace=True)
data['country']=data['country'].fillna(data['country'].mode()[0])
data['date_added']=data['date_added'].ffill()
data['rating']=data['rating'].fillna(data['rating'].mode()[0])
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data['date_added']=pd.to_datetime(data['date_added'], format= '%B %d, %Y')
```

```python
data['date_added']
```

```python
data['rating'].unique()
```

```python
data['rating']=data['rating'].replace({
    "NOT_RATE":"UNRATED",
    "AGES_16_":"16+",
    "16":"16+",
    "AGES_18_":"18+"
})
```

```python
data['rating'].unique()
```

```python
data['type'].value_counts()
```

```python
data.head()
```

```python
data.rename(columns={'listed_in':'categories'},inplace=True)
data
```

```python
data['duration_min']=data['duration'].apply(lambda x:int(x.split()[0]) if 'min' in x else 0)
```

```python
data['duration_season']=data['duration'].apply(lambda x:int(x.split()[0]) if 'Seasons' in x else 0)
```
